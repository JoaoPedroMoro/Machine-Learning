### Pré-processamento

Nesta seção, as funções da etapa de pré-processamento dos dados devem ser implementadas e aplicadas (se necessário).

```python
if __name__ == '__main__':
    dados_clinicos = pd.read_csv( os.path.join("dataset\dataset", 'clinical_data.csv'), sep=',') 
    # Tratamento de espaços em branco desnecessários
    print("Removendo espaços no início ou fim dos nomes de atributos e nos valores.")
    print("Fazendo uma pesquina e uma análise rápida, foi descoberto que as colunas 'Sex ' e 'Race            ' dos dados clínicos são atributos irrelevantes para o diagnóstico da doença.\nDessa forma, serão eliminados do dataframe.")
    dados_clinicos = dados_clinicos.drop(columns=['Sex ','Race            '])
    print("Irá ser retirado todos os espaços em branco depois dos valores dos dataframes.")
    # Remoção de espaço dos nomes das colunas
    dados_clinicos.columns = dados_clinicos.columns.str.strip()
    dados_mutacao_genetica.columns = dados_mutacao_genetica.columns.str.strip()
    dados_expressao_genetica.columns = dados_expressao_genetica.columns.str.strip()
    treinamento.columns = treinamento.columns.str.strip()
    teste.columns = teste.columns.str.strip()
    # Remoção de espaço no valores
    dados_clinicos = dados_clinicos.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    dados_mutacao_genetica = dados_mutacao_genetica.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    dados_expressao_genetica = dados_expressao_genetica.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    dados_clinicos = dados_clinicos.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    treinamento = treinamento.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    teste = teste.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    
    # Tratamento de valores ausentes
    print("\nAgora irá ser verificado em todos os dataframes se há atributos com valores ausentes.")
    ausDC = dados_clinicos.eq('').any(axis=1)
    ausMG = dados_mutacao_genetica.eq('').any(axis=1)
    ausEG = dados_expressao_genetica.eq('').any(axis=1)
    ausTest = teste.eq('').any(axis=1)
    ausTrain = treinamento.eq('').any(axis=1)
    print("Apenas os dados clínicos possuem valores ausentes, por isso devemos tratá-los.")
    
    print("Como todos os atributos faltantes são do tipo Object, vamos tratá-los da mesma forma, preencher esses valores com a média do atributo.")
    dados_clinicos_obj = dados_clinicos.select_dtypes(include='object') ## Todos os atributos de valores ausentes são do tipo Object, então vamos selecionar todos eles para tratá-los.
    dados_clinicos_obj.replace('', np.nan, inplace=True) ## Agora para facilitar, iremos substituir por NaN
    mediaAtributos_obj = dados_clinicos_obj.apply(lambda x: pd.to_numeric(x, errors='coerce')).mean() ## Vamos converte-los todos para o tipo numerico e inserir a media do atributo no seu lugar.
    dados_clinicos.fillna(value=mediaAtributos_obj, inplace=True) ## Vamos substituir agora no dataframe de dados clínicos os valores ausentes.
    print("Todos os atributos faltantes já foram corrigidos.")
    
    # Tratamento de amostras redundantes ou inconsistentes
    print("\nAgora será verificado se há amostras redundantes ou inconsistentes em todos os dataframes e, se tiver, serão tratadas essas amostras.")
    dcDp = dados_clinicos[dados_clinicos.duplicated(subset=['Sample ID'],keep=False)]
    if len(dcDp)>0:
        print('Existem amostras inconsistentes ou redundantes nos dados clínicos!')
        dados_clinicos = dados_clinicos.drop_duplicates()
    mgDp = dados_mutacao_genetica[dados_mutacao_genetica.duplicated(subset=['Sample ID'],keep=False)]
    if len(mgDp)>0:
        print('Existem amostras inconsistentes ou redundantes nos dados de mutação genética!')
        dados_mutacao_genetica = dados_mutacao_genetica.drop_duplicates()
    egDp = dados_expressao_genetica[dados_expressao_genetica.duplicated(subset=['Sample ID'],keep=False)]
    if len(egDp)>0:
        print('Existem amostras inconsistentes ou redundantes nos dados de expressão genética!')
        dados_expressao_genetica = dados_expressao_genetica.drop_duplicates()
    testDp = teste[teste.duplicated(subset=['Sample ID'],keep=False)]
    if len(testDp)>0:
        print('Existem amostras inconsistentes ou redundantes no conjunto de dados de teste!')
        teste = teste.drop_duplicates()
    trainDp = treinamento[treinamento.duplicated(subset=['Sample ID'],keep=False)]
    if len(trainDp)>0:
        print('Existem amostras inconsistentes ou redundantes no conjunto de dados de treinamento!')
        treinamento = treinamento.drop_duplicates()
    print("Verificando se ficou alguma amostra inconsistente ou duplicado...")
    dcDp = dados_clinicos[dados_clinicos.duplicated(subset=['Sample ID'],keep=False)]
    mgDp = dados_mutacao_genetica[dados_mutacao_genetica.duplicated(subset=['Sample ID'],keep=False)]
    egDp = dados_expressao_genetica[dados_expressao_genetica.duplicated(subset=['Sample ID'],keep=False)]
    testDp = teste[teste.duplicated(subset=['Sample ID'],keep=False)]
    trainDp = treinamento[treinamento.duplicated(subset=['Sample ID'],keep=False)]
    if len(dcDp)>0 or len(mgDp)>0 or len(egDp)>0 or len(testDp)>0 or len(trainDp)>0:
        print("Ainda há amostras redundantes ou inconsistentes, tratando-as novamente...")
        dados_clinicos = dados_clinicos.drop_duplicates()
        dados_mutacao_genetica = dados_mutacao_genetica.drop_duplicates()
        dados_expressao_genetica = dados_expressao_genetica.drop_duplicates()
        teste = teste.drop_duplicates()
        treinamento = treinamento.drop_duplicates()
    else:
        print("Não existem mais amostras inconsistentes ou redundantes em nenhum dataframe!")
        
    # Normalização dos dados e verificação da existência de outliers
    X = dados_clinicos.iloc[:, 2:6].values
    display(dados_clinicos.iloc[:, 2:6])
    X = X.astype('U')  # converte para tipo string
    X = np.char.replace(X, ',', '.')  # substitui as vírgulas pelos pontos
    X = X.astype(float)  # converte para tipo float
    print(X)
    mu = X.mean(axis=0)
    # Calculando o desvio padrão dos atributos em sigma
    sigma = X.std(axis=0, ddof = 1) # axis = 0 para contabilizar as colunas
    # Realizando a normalização
    X_norm = (X-mu)/sigma
    dados_clinicos.iloc[:, 2:6] = X_norm
    # Detecção de outliers
    print("\nComo os dataframes relacionados a genes são de presença ou não de mutação e variação do gene no paciente, não existem outliers para esses dois dataframes.")
    print("O data frame de teste possui apenas os ids dos pacientes, então também não possui outliers.")
    print("Dessa forma, iremos verificar pelos gráficos se existem outliers nos dados clínicos e no conjunto de treinamento:")
    dados_clinicos.boxplot(figsize=(15,7))    
    plt.show()
    treinamento.boxplot(figsize=(15,7))
    plt.show()
    print("Pelos gráficos, apenas os dados clínicos precisam ser normalizados.")
    # for atributo in dados_clinicos.columns[2:6]:
    #     # Removendo os valores fora do intervalo [-1, 1]
    #     dados_clinicos = dados_clinicos[~((dados_clinicos[atributo] < -1) | (dados_clinicos[atributo] > 1))]
    print("Vamos verificar se há ainda mais algum outlier nos dados clínicos.")
    dados_clinicos.boxplot(figsize=(15,7))
    plt.show()
    print("Como se pode ver pelo gráfico, não existem outliers para serem tratados.")
    
    print("\nDepois de todas essas operações, foi feito o pré-processamento de dados!")
```
